```python
# Read the data from a text file
csv1 = pd.read_csv('loop1.CT.txt', delimiter='\t', header=None, skipinitialspace=True)
csv2 = pd.read_csv('loop2.CT.txt', delimiter='\t', header=None)
csv3 = pd.read_csv('loop3.CT.txt', delimiter='\t', header=None)
csv4 = pd.read_csv('loop4.CT.txt', delimiter='\t', header=None)
```

```python
csv1=csv1.iloc[:,1:]
csv2=csv2.iloc[:,1:]
csv3=csv3.iloc[:,1:]
csv4=csv4.iloc[:,1:]
```

```python
csv1.mean(axis=1).to_csv('loop1.txt',index=False)
csv2.mean(axis=1).to_csv('loop2.txt',index=False)
csv3.mean(axis=1).to_csv('loop3.txt',index=False)
csv4.mean(axis=1).to_csv('loop4.txt',index=False)
```

```python
cnt = 0
with open('loop1.txt', 'w') as f:
    for i in csv1.mean(axis=1):
        
        f.write("%10.6f" % i)
        cnt = cnt + 1
        
        if cnt % 8 == 0:
            f.write("\n")
```
